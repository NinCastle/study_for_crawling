# Text Mining Information
아래의 내용은 IBM에서 한글로 자동번역한 느낌이 나서 조금 수정해서 정리.

데이터 분석의 오늘날의 화두는 "정보를 어떻게 수집, 탐색하고 활용할 수 있나?"이다.<br>
<br>
**Text Mining : **<br>
작성자가 개념을 표현하기 위해 사용한 정확한 단어와 용어를 모르더라도 주요 개념과 테마를 캡처하고 숨겨진 관계와 경향을 발견하기 위해 텍스트 자료 콜렉션을 분석하는 프로세스이다.<br>
<br>
텍스트 마이닝은 정보검색과는 상당히 다르기는 하지만 종종 혼동되기도 한다.<br>

## 텍스트 마이닝 및 데이터 마이닝

Text의 각 기사마다 언어학적 기반 Text Mining은 개념 지수뿐만 아니라 개념에 대한 정보를 리턴한다.<br>
<br>
구조화된 정보는 다른 데이터 소스와 결합하여 아래와 같은 질문을 처리한다.<br>
* 어떤 개념이 함께 발생하나?
* 그 밖에 어디에 링크 되어있나?
* 추출된 정보로부터 어떤 사위 수준 범주를 작성할 수 있나?
* 개념 또는 범주가 예상하는 것은 무엇인가?
* 개념 또는 범주가 작동을 어떻게 예상합니까?

텍스트 마이닝을 데이터 마이닝과 결합하면서 구조화된 또는 구조화되지 않은 데이터만 사용한 것보다 더 많은 통찰력을 제공한다.<br>
이 프로세스는 다음 단계를 포함한다.<br>
1. **마이닝할 텍스트를 식별한다.**<br>
텍스트 마이닝을 할 준비를 하고 텍스트가 여러 파일에 존재하면 하나의 파일로 저장한다.<br>
데이터 베이스의 경우 텍스트를 포함하는 필드를 판별한다.<br>
<br>
2. **텍스트를 마이닝하고 구조화된 데이터를 추출한다.**<br>
텍스트 마이닝 알고리즘을 소스 텍스트에 적용한다.<br>
<br>
3. **개념 및 범주 모델을 작성한다.**<br>
주요 개념을 식별하고 범주를 정한다.<br>
구조화되지 않은 데이터로 부터 리턴된 개념 수는 일반적으로 매우 크다.<br>
스코어링을 위해 최상의 개념과 범주를 식별한다.<br>
<br>
4. **구조화된 데이터를 분석**<br>
군집, 분류 및 예측 모델링과 같은 일반적인 데이터 마이닝 기술을 사용하여 개념 간의 관계를 발견한다.<br>
추출된 개념을 다른 구조화된 데이터와 병합하여 개념을 기반으로 추가로 동작을 예측한다.<br>

## 텍스트 분석 및 범주화
**질적 분석의 양식으로 된 텍스트 분석 : **<br> 
텍스트에 포함된 주요 아이디어와 개념이 적합한 개수의 범주로 그룹화 될 수 있도록 텍스트로 부터 유용한 정보를 추출하는 것을 말한다.<br>
텍스트 분석은 접근 방법이 다소 다르더라도 모든 유형과 텍스트 길이에서 수행될 수 있다.<br>
<br>
짧은 레코드 또는 문서가 가장 쉽게 범주화된다.<br>
이것들은 복잡하지 않고 일반적으로 애매한 단어나 반응이 적기 때문이다.<br>
예를 들어,<br>
짧은 개방형 설문 질문에서 사람들에게 가장 좋아하는 휴가 활동을 뽑으라고 할때.<br>
1. 해변에 가기
2. 국립공원 방문
3. 아무것도 안하기

등과 같은 짧은 답변을 여러개 예상할 수 있다.<br>
반면 더 긴 개방형 반응은 복잡하고 길 수 있으며,<br>
* 반응자의 교육상태
* 반응자의 동기
* 반응자의 설문지 작성시간

등과 같은 것들이 영향을 미친다.<br>
설문에서 사람들에게 자신의 정치적 신념에 대해 얘기해 달라고 묻거나 정치에 대한 블로그 피드를 가지도록 요청한다면 모든 종류의 문제와 위치에 대해 다소 긴 설명을 예측할 수 있다.<br>
<br>

## 언어학적 처리와  NLP
구조화되지 않은 모든 텍스트 데이터 관리의 주요 문제점은 컴퓨터가 이해할 수 있도록 텍스트를 쓰기 위한 표준 규칙이 없다는 점이다.<br>
언어의 의미는 모든 문서 및 텍스트마다 의미가 다르다.<br>
이러한 구조화되지 않은 데이터를 정확하게 검색하고 조직하는 유일한 방법은 **언어를 분석하고 해당 의미를 발견**하는것이다.<br>
구조화되지 않은 정보로 부터 개념 추출을 위한 여러 개의 자동화된 방법이 있다.<br>
위와 같은 접근 방법은 언어학적 과 비언어학적 두가지 종류로 구분 할 수 있다.<br>
