```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator

def make_ImageColoredWordcloud(text, img_matrix):
    wc = WordCloud(background_color="white", max_words=2000, mask=img_matrix, max_font_size=40, random_state=42)
    
    #generate word cloud
    wc.generate(text)
    f = plt.figure(text, figsize=(30, 40))
    plt.imshow(wc.recolor(color_func=ImageColorGenerator(img_matrix)), interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
url = "http://ipakwheels.com/wp-content/uploads/2017/12/dc-avanti.jpg"
img_matrix = read_img_from_url_and_return_matrix(url)
text = requests.get("https://en.wikipedia.org/wiki/Car").text
make_ImageColoredWordcloud(text, img_matrix)
```
