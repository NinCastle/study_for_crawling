# xml 크롤링

```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse
```

```python
web_url = "https://data.kma.go.kr/OPEN_API/AWSM/2015/09/XML/month.xml"
with urllib.request.urlopen(web_url) as response:
    xml = response.read()
    soup = BeautifulSoup(xml, 'xml')
soup
```
