## String to XML

```python
def string_to_xml(text):
    root = ET.fromstring(text)
    fil_text = []
    for n in root.findall('info'):
        lo = "지역"
        tem = "평균기온"
        wat = "월 강수량"
        dict_to = {}
        try:
            dict_to[lo] = str(n.find('stn_ko').text)
        except ValueError:
            break
        try:
            dict_to[tem] = float(n.find('avg_ta_day').text)
        except ValueError:
            break
        try:
            dict_to[wat] = float(n.find('sum_rn_day').text)
        except ValueError:
            break
        fil_text.append(dict_to)
    return fil_text
```

## XML에서 크롤링하기

```python
def xml_crawling(i, j, url):
    with urllib.request.urlopen(url) as response:
        xml = response.read()
        soup = BeautifulSoup(xml, 'xml')
        soup_total = string_to_xml(str(soup))
        print("년도 : ",i, " 월 :", j)
    return soup_total
```

## 평균 함수

```python
def make_average(value, str_value):
    total = 0
    
    for i in range(len(value)):
        total = total + value[i][str_value]
#     print("총 값 : ",total)
    return total/len(value)
```
